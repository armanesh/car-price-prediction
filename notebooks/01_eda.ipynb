# Exploratory Data Analysis — Used Car Prices

This notebook explores the Craigslist vehicles dataset to understand distributions, correlations, and data quality issues before modeling.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from src.data.load import load_raw
from src.data.preprocess import run as preprocess

sns.set_theme(style='whitegrid')
plt.rcParams['figure.figsize'] = (12, 5)
```

## 1. Load Raw Data

```python
raw = load_raw()
print(raw.shape)
raw.head()
```

## 2. Missing Values

```python
missing = raw.isnull().mean().sort_values(ascending=False) * 100
missing[missing > 0].plot(kind='bar', title='Missing Values (%)')
plt.tight_layout()
plt.show()
```

## 3. Price Distribution

```python
df = preprocess(raw)

fig, axes = plt.subplots(1, 2)
axes[0].hist(df['price'], bins=80, color='steelblue', edgecolor='white')
axes[0].set_title('Price Distribution (cleaned)')
axes[0].set_xlabel('Price ($)')

axes[1].hist(np.log1p(df['price']), bins=80, color='steelblue', edgecolor='white')
axes[1].set_title('log(Price) Distribution')
axes[1].set_xlabel('log(Price)')

plt.tight_layout()
plt.show()
```

## 4. Price vs Vehicle Age & Odometer

```python
fig, axes = plt.subplots(1, 2)

sample = df.sample(10_000, random_state=42)
axes[0].scatter(sample['vehicle_age'], sample['price'], alpha=0.15, s=5, color='steelblue')
axes[0].set_xlabel('Vehicle Age (years)')
axes[0].set_ylabel('Price ($)')
axes[0].set_title('Price vs Age')

axes[1].scatter(sample['odometer'], sample['price'], alpha=0.15, s=5, color='coral')
axes[1].set_xlabel('Odometer (miles)')
axes[1].set_ylabel('Price ($)')
axes[1].set_title('Price vs Odometer')

plt.tight_layout()
plt.show()
```

## 5. Median Price by Manufacturer (Top 15)

```python
top_makes = df.groupby('manufacturer')['price'].median().sort_values(ascending=False).head(15)
top_makes.plot(kind='bar', color='steelblue', title='Median Price by Manufacturer')
plt.ylabel('Median Price ($)')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

## 6. Correlation Matrix (Numeric Features)

```python
num_cols = ['price', 'vehicle_age', 'odometer', 'age_x_odometer']
corr = df[num_cols].corr()
sns.heatmap(corr, annot=True, fmt='.2f', cmap='coolwarm', center=0)
plt.title('Correlation Matrix')
plt.tight_layout()
plt.show()
```
